```python
#outliers using Z score
def detech_outliers(ohed, threshold =3):

    ohed = np.array(ohed, dtype=np.float64)
    mean=np.mean(ohed)
    std=np.std(ohed)

    
    z_scores=(ohed-mean)/std
    outliers = ohed[np.abs(z_scores) > threshold]    
    return outliers
```
