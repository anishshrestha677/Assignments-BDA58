##### IMDb Movie Reviews: Scraping & Cleaning the movie ratings

```python
from bs4 import BeautifulSoup 
import requests 
import pandas as pd
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.; Win64; x64) AppleWeKit/537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
url = "https://www.imdb.com/chart/top/"
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
page
```

```python
movies_list =[]
for movie_container in soup.find_all('li', class_= 'ipc-metadata-list-summary-item sc-10233bc-0 TwzGn cli-parent'):
    movie_title = movie_container.find('h3').text.split('.')[1]
    movie_year = movie_container.find('div', class_='sc-b189961a-7 btCcOY cli-title-metadata').span.text
    #movie_run_time = movie_container.find('span', class_='sc-b189961a-8 hCbzGp cli-title-metadata-item')[1].text
    movie_rating = movie_container.find('div', class_= 'sc-e2dbc1a3-0 jeHPdh sc-b189961a-2 bglYHz cli-ratings-container').span.text

    movies_list.append({
        'Movie Title': movie_title,
        'Release Date': movie_year,
        #'Running Time': movie_run_time,
        'Movie Genre': movie_rating
    })
    
movies_list
  
    
```

```python
df = pd.DataFrame(movies_list)
df
```
