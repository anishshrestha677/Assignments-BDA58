##### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("Mall_Customers.csv")
df
```

###### Feature Engineering
###### Male and Female to binanry form

```python
one_hot_encoded_data = pd.get_dummies(df, columns = ['Genre'], dtype=int)
one_hot_encoded_data
```

###### Processing Data
###### taking the standard deviation 

```python
scaler = StandardScaler()
ohe_scaled = scaler.fit_transform(one_hot_encoded_data)
print(ohe_scaled)
```

### k Means Clustering
###### using K Mean Clustering

```python
kmeans = KMeans(n_clusters=4)
# when we use k mean cluster have to pass two parameters
# n_cluster (how many clusters we are looking for)

kmeans.fit(ohe_scaled)
#passing the data to fit it in the model

clusters = kmeans.labels_
print(f"Cluster assignment: {clusters}")
#assign all datas to clusters
```

##### Visualizing Clusters

```python
one_hot_encoded_data['Cluster'] = clusters

plt.figure(figsize=(20,10))
plt.scatter(one_hot_encoded_data['Annual Income (k$)'], 
            one_hot_encoded_data['Spending Score (1-100)'],
            one_hot_encoded_data['Age'],
            c=one_hot_encoded_data['Cluster'],
            cmap='viridis')

plt.title('Target Customer Compass')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

##### The clusters represet group of customers with similar spending habits. Targetting each customer group with lucrative deals will be easier

### Hierarchical Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage

data = list(zip(one_hot_encoded_data['Annual Income (k$)'], 
                one_hot_encoded_data['Spending Score (1-100)'],
                one_hot_encoded_data['Genre_Female'],
                one_hot_encoded_data['Age'],
                one_hot_encoded_data['Genre_Male']))

linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show

```
