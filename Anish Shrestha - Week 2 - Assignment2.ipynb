```python
#Parse the HTML with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
soup
```

```python
movies_list = []
for movie_container in soup.find_all('div', class_='movie-container'): 
    # using iterator movie_container, iterator will be using to define the path to extract info.
    #we are finding all div and inside div going to class='movie-container' to gain more inside info
    movie_title = movie_container.find('h1').text
    #inside h1 tag we have movie title with .text only getting the text
    movie_release = movie_container.find_all('p')[0].text.split(': ')[1]
    movie_genre = movie_container.find_all('p')[1].text.split(': ')[1]
    movie_director = movie_container.find_all('p')[2].text.split(': ')[1]
    movie_cast = movie_container.find_all('p')[3].text.split(': ')[1]

    movies_list.append({
        'Movie Title': movie_title,
        'Release Date': movie_release,
        'Movie Genre': movie_genre,
        'Movie Director': movie_director,
        'Movie Cast': movie_cast
    })
    
movies_list
```

```python
#create a DataFrame
df = pd.DataFrame(movies_list)
df
```

```python
# save the data to csv file
df.to_csv('Movie list.csv', index=False)
```
